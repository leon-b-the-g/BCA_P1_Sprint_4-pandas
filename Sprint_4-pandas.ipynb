```python
###Pandas excercises###
import pandas as pd
```

```python
#General info about dataframe
movie_df = pd.read_csv('movies_2.csv') # Gathering data
movie_df.head(5)

movie_df.info()
movie_df.shape

```

```python
#Find the amount of NaNs
Nancount = movie_df.isnull().sum()
print("These are the amount of NaNs:", Nancount)
```

```python
#remove Summary column
no_summary_df = movie_df.drop("Summary",axis = 1)

no_summary_df.head(5)
```

```python
#Rename column headers (features) to lower case with _ seperation

#take a list of column header and format them
column_headers = no_summary_df.columns.tolist()
column_headers_new = []
for elements in column_headers:
    nospace = elements.replace(" ","_")
    column_headers_new.append(nospace.lower())

print("This is new column headers: ",column_headers_new)

#provide a dictionary saying which column labels need to be replaced by what new column labels
guide_dict = {}
counter = 0
for old_header in column_headers:
    guide_dict[old_header] = column_headers_new[counter]
    counter += 1

print("This is guide dict: ",guide_dict)

movie_df_newheaders = no_summary_df.rename(columns=guide_dict)
movie_df_newheaders.tail(5)
```

```python
#Drop last row since it has NaN values
clean_df = movie_df_newheaders.drop(index = 615, axis = 0)
clean_df.tail(5)
```

```python
#Replace missing NaN values in Rating column with median rating
clean_df["rating"].fillna(value=clean_df["rating"].median(), inplace=True)
clean_df["rating_count"].fillna(value=clean_df["rating_count"].median(), inplace=True)

#clean_df.tail(5)
clean_df.dtypes
```

```python
#Remove duplicates:
no_dupes = clean_df.drop_duplicates(subset = ['title'], inplace= False)
no_dupes.head(5)
```

```python
#Convert Budget into integer type (use astype() method)
#Convert Gross into integer type
#Convert Release Date into date type (use to_datetime method)

change_dtypes = no_dupes.astype({"budget":"int32","gross":"int32"})


change_dtypes["release_date"] = pd.to_datetime(change_dtypes["release_date"])
change_dtypes.info()
```

```python
#Answer the following questions:

#show the movies with more than 7 in Rating & greater than 50 million Gross
#pull movie names, rating and gross, then filter resulting df
name_rating_gross = change_dtypes.loc[:,['title','rating','gross']]
name_rating_gross.head(5)

only_the_best = name_rating_gross.loc[name_rating_gross['rating'] > 7]
only_the_greatest = only_the_best.loc[only_the_best['gross'] > 50000000]

only_the_best.info()
only_the_greatest.info()

```

```python
#show the movies with more than 7 in Rating & greater than 50 million Gross & with Parental guidance as MPAA Rating
#filter for >7 rating, >50 mil gross and MPAA == PG

mpaa_rating_gross = change_dtypes.loc[:,['title','mpaa_rating','gross','rating']]
#mpaa_rating_gross_dtypesadjusted = mpaa_rating_gross.astype(({"gross":"float64"}))  # converting rating back, screwing up operators 
#mpaa_rating_gross_dtypesadjusted.info()
filter_1 = mpaa_rating_gross.loc[mpaa_rating_gross['rating']> 7]
filter_1.info()
filter_2 = filter_1.loc[filter_1['gross']> 50000000]

filtered_df_mpaa_rating_gross = filter_2.loc[filter_2['mpaa_rating'] == 'PG']

filtered_df_mpaa_rating_gross.info()
```

```python
#most and least expensive Genre (take an average of all Budget measures grouped by Genre - use groupBy() method)

all_genres= clean_df.groupby(by="genre").mean("budget")
print(all_genres)
print("Most expensive genre: Fantasy, with: $ ",round(all_genres["budget"].max()))

print("Least expensive genre: Horror, with: $ ",round(all_genres["budget"].min()))

```

```python
#which Genre is favoured the most by the people?
#group by genre, find average rating, find best rated genre
all_genres_ratings= clean_df.groupby(by="genre").mean("rating")
print(all_genres_ratings)
best_rated = all_genres_ratings["rating"].max(skipna=True)
print(f"The best rated Genre is History, with a mean rating of: ", best_rated)
```
